## Notebook 03 - Data Merging
This notebook merges the cleaned game logs data (from Notebook 02) with:

- [NBA 2024-2025 Per Game Stats](https://www.basketball-reference.com/leagues/NBA_2025_per_game.html) (copied and pasted into `.txt` file)
- [NBA 2024-2025 Shooting Stats](https://www.basketball-reference.com/leagues/NBA_2025_shooting.html) (copied and pasted into `.txt` file)

The goal is to add **three additional columns** to the cleaned game logs data:

- position
- age
- average shot distance (season-long aggregate)

```python
# Import libraries
import pandas as pd
import numpy as np
```

## Load Data

```python
# Load cleaned game logs (from Notebook 02)
cleaned = pd.read_csv(r"C:\Users\dylan\OneDrive\Documents\Portfolio_Projects\project10_NBA_2025_game_logs\03_python_outputs\01_cleaning\NBA_2024_25_game_logs.csv")

print(f"Loaded: {cleaned.shape[0]:,} rows | {cleaned["player_name"].nunique():,} players")
```

```python
# Load Per Game Stats .txt file
per_game = pd.read_csv(r"C:\Users\dylan\OneDrive\Documents\Portfolio_Projects\project10_NBA_2025_game_logs\01_original_data\to_merge\NBA_league_leaders_per_game.txt")

print(f"Loaded: {per_game.shape[0]:,} rows | {per_game["player_name"].nunique():,} players")
```

```python
# Load Shooting Stats .txt file
shooting_stats = pd.read_csv(r"C:\Users\dylan\OneDrive\Documents\Portfolio_Projects\project10_NBA_2025_game_logs\01_original_data\to_merge\NBA_league_leaders_shooting.txt")

print(f"Loaded: {shooting_stats.shape[0]:,} rows | {shooting_stats["player_name"].nunique():,} players")
```

___
## Merge "Per Game Stats"

```python
# Show all columns
per_game.columns
```

> Only `age` and `pos` are needed.

```python
# Keep only `age` and `pos` columns (and `player_name` as linking field)
per_game_needed_cols = per_game[["player_name", "age", "pos"]].copy()

# Rename `pos` to `position` for clarity
per_game_needed_cols = per_game_needed_cols.rename(columns={"pos": "position"})
```

```python
# Merge these 2 columns back into the cleaned game logs data
cleaned_merged = cleaned.merge(per_game_needed_cols, on="player_name", how="left")
```

## Merge "Shooting Stats"

```python
# Show all columns
shooting_stats.columns
```

> Only `distance` is needed.

```python
# Keep only `distance` column (and `player_name` as linking field)
shooting_stats_needed_cols = shooting_stats[["player_name", "distance"]].copy()
```

```python
# Merge this column back into the cleaned game logs data
merged_final = cleaned_merged.merge(shooting_stats_needed_cols, on="player_name", how="left")
```

___
## Sanity Checks
